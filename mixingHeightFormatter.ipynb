## Format files

```python
files = glob.glob('../mixing-heights/KABQ/KABQ/*')
```

```python
df = pd.read_csv(files[0])

for file in files[1:]:
    df2 = pd.read_csv(file)
    df = df.append(df2)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df = df.sort_values(by='Date')
df['Date'] = df['Date'].dt.strftime('%y%m%d/0000')
df
```

```python
to_save = df.loc[:,['Date','SfcThV_Hgt']]
to_save.rename(columns={'Date':'date','SfcThV_Hgt':'val'},inplace=True)
to_save['val'] = pd.to_numeric(to_save['val'],downcast='float')
to_save.to_csv('abq-mix_sfc-filtered.csv',index=False)
```

```python
mixparms = [['SfcThV_Hgt','mix_sfc'],
            ['SfcThV+1K_Hgt','mix_sfc_1k'],
            ['500mThV_Hgt','mix_500m'],
            ['SfcThV_Pres','mix_sfc_pres'],
            ['SfcThV+1K_Pres','mix_sfc_1k_pres'],
            ['500mThV_Pres','mix_500m_pres']]

stn = 'abq'

for parm in mixparms:
    to_save = df.loc[:,['Date',parm[0]]]
    to_save.rename(columns={'Date':'date',parm[0]:'val'},inplace=True)
    to_save['val'] = pd.to_numeric(to_save['val'],downcast='float')
    to_save.to_csv(f'./datafiles/{stn}/{stn}-{parm[1]}-filtered.csv',index=False)
```
